```python
# 인터페이스 실험

h = float(input('드랍박스와 time plate 간의 거리 h를 입력해주세요(단위:m): '))
r = float(input('공의 반지름을 입력해주세요(단위:m): '))
t = float(input('공의 낙하시간을 입력해주세요: '))

def gravitational_acceleration(h, r, t) :
    real_h = h - 2*r
    g = round(2 * real_h / pow(t, 2), 5)
    real_g = 9.80665
    Error_rate = round(abs(real_g - g) / real_g * 100, 5)

    fin = '\n결과 = g: {}, 오차율: {}(단, g = 9.80665m/s^2)'.format(g, Error_rate)
    return fin

gravitational_acceleration(h, r, t)
```

```python
# 에어트랙 중력가속도 실험

h = float(input('받침목의 높이 h를 입력해주세요(단위 : m): '))
D = float(input('포토게이트 사이의 거리 D를 입력해주세요(단위 : m): '))
t = float(input('공이 D만큼 이동하는데 걸린 시간의 평균을 입력해주세요: '))

def gravitational_acceleration_airtrack(h, D, t) :
    sine_zeta = h/D
    g = round(2 * D / (pow(t, 2)*sine_zeta), 5)
    real_g = 9.80665
    Error_rate = round(abs(real_g - g) / real_g * 100, 5)

    fin = '결과 = g: {}, 오차율: {}(단, g = 9.80665m/s^2)'.format(g, Error_rate)
    return fin

gravitational_acceleration_airtrack(h,D,t)
```

```python
# 포물체 운동 실험

# 실험 A
t = float(input('t의 평균값을 입력하세요: '))
L = float(input('포토게이트 헤드 사이의 거리를 입력하세요(단위 : m): '))
v0 = round(L/t, 5)
fin = 'v0 = ' + str(v0)
print(fin)

# 실험 B

def parabola(zeta, x_average, t_average) :
    import math
    g = 9.80665
    a = math.sin(math.radians(zeta))
    t_theory = round(2*v0*a / g, 4)
    x_theory = round(v0 * math.cos(math.radians(zeta)) * t_theory, 4)
    Error_rate_t = round(abs(t_average - t_theory) / t_theory, 4) * 100
    Error_rate_x = round(abs(x_average - x_theory) / x_theory, 4) * 100
    y = round(pow(v0 * a, 2)/(2*g), 5)
    return '결과 = x의 이론값: {}, t의 이론값: {}, x의 오차율: {}, t의 오차율: {}, 최고도달높이 y: {}'.format(x_theory, t_theory,
                                                                                 Error_rate_x, Error_rate_t, y)
k = 0
while k <= 5 : 
    zeta = int(input('발사대의 각도를 입력하세요: '))
    x_average = float(input('수평이동거리의 평균을 입력하세요(단위 : m): '))
    t_average = float(input('걸린 시간의 평균을 입력하세요: '))
    print(parabola(zeta, x_average, t_average))
    k += 1

```

```python
# 역학적 에너지 보존 실험
# 항상 일정
d = float(input("에어트랙 받침대 사이의 거리 d를 입력하세요(단위 : m): "))
L = float(input("플래그의 길이 L을 입력하세요(단위 : m): "))
g = 9.80665

# 변할 수 있음
m = float(input("글라이더의 질량 m을 입력하세요(단위 : kg): "))
h = float(input("받침목의 높이 h를 입력하세요(단위 : m)"))
D = float(input("포토게이트 사이의 거리 D를 입력하세요: "))
t1 = float(input("첫번째 포토게이트를 지날 때 걸린 시간을 입력하세요: "))
t2 = float(input("두번째 포토게이트를 지날 때 걸린 시간을 입력하세요: "))
def conservation_of_mechanical_energy(t1, t2, D, m, h) : 
    v1 = round(L/t1, 4)
    v2 = round(L/t2, 4)
    Eki = round(0.5*m*pow(v1, 2), 4)
    Ekf = round(0.5*m*pow(v2, 2), 4)
    delta_Ek = round(Ekf-Eki, 4)
    delta_Ep = round(m*g*D*h/d, 4)
    error_rate = round(abs(delta_Ek - delta_Ep) / delta_Ep * 100, 4)
    return '결과 = v1: {}, v2: {}, 처음 Ek: {}, 나중 Ek: {}, Ek 변화량: {}, Ep 변화량: {}, 오차율: {}'.format(v1, v2, Eki, Ekf, delta_Ek,
                                                                                         delta_Ep, error_rate)
    
conservation_of_mechanical_energy(t1, t2, D, m, h)
```

```python
# 운동량 보존 실험

print("만약 측정되지 않는 값이라면 '0'을 입력하세요\n")

# 일정한 것
L1 = float(input("글라이더1의 플래그 길이 L1을 입력하세요(단위:m): "))
L2 = float(input("글라이더2의 플래그 길이 L2를 입력하세요(단위:m): "))

# 변하는 것
m1 = float(input("글라이더1의 질량 m1을 입력하세요(단위:kg): "))
m2 = float(input("글라이더2의 질량 m2를 입력하세요(단위:kg): "))
t1i = float(input("충돌 전 글라이더1이 포토게이트를 통과한 시간을 입력하세요: "))
t2i = float(input("충돌 전 글라이더2가 포토게이트를 통과한 시간을 입력하세요: "))
t1f = float(input("충돌 후 글라이더1이 포토게이트를 통과한 시간을 입력하세요: "))
t2f = float(input("충돌 후 글라이더2가 포토게이트를 통과한 시간을 입력하세요: "))
print("\n글라이더의 이동 방향을 '+'와 '-'로 입력해주세요. 만약 움직이지 않았다면 '0'을 입력해주세요.\n")
direct_1i = input("충돌 전 글라이더1의 이동방향: ")
direct_2i = input("충돌 전 글라이더2의 이동방향: ")
direct_1f = input("충돌 후 글라이더1의 이동방향: ")
direct_2f = input("충돌 후 글라이더2의 이동방향: ")
direct = [direct_1i, direct_2i, direct_1f, direct_2f]

def conservation_of_momentum(m1, m2, t1i, t2i, t1f, t2f, direct) : 
    velocity1 = [t1i, t1f]
    velocity2 = [t2i, t2f]
    velocity = [velocity1, velocity2]
    for i in velocity : 
        for j in i : 
            if j == 0 : 
                i[i.index(j)] = 0
            else : 
                if i == velocity1 : 
                    i[i.index(j)] = round(L1/j, 4)
                else : 
                    i[i.index(j)] = round(L2/j, 4)
    
    
    v1i = velocity[0][0]
    v2i = velocity[1][0]
    v1f = velocity[0][1]
    v2f = velocity[1][1]
    p1i = round(m1*v1i, 4)
    p1f = round(m1*v1f, 4)
    p2i = round(m2*v2i, 4)
    p2f = round(m2*v2f, 4)
    momentum = [p1i, p2i, p1f, p2f]
    for i in range(4) : 
        if direct[i] == '-' : 
            momentum[i] *= -1
    pi = round(momentum[0] + momentum[1], 4)
    pf = round(momentum[2] + momentum[3], 4)
    error_rate = round(abs(pi-pf)/pi*100, 4)
    return '결과 = v1i: {}, v1f: {}, v2i: {}, v2f: {}, P1i: {}, P1f: {}, P2i: {}, P2f: {}, Pi: {}, Pf: {}, 오차율: {}'.format(v1i,v1f,v2i,v2f,
                                                                                                                   p1i, p1f,p2i,p2f,pi,pf,error_rate)

conservation_of_momentum(m1, m2, t1i, t2i, t1f, t2f, direct)
```

```python
# 탄성 및 비탄성충돌 실험
# 운동에너지
def kinetic_energy(m, v) : 
    return round(0.5*m*pow(v,2),4)
# 오차율
def error_rate(theory, ex) : 
    return round(abs(theory-ex)/theory*100, 4)

# 실험 A 함수
def elastic_collision(m1, m2, t1i, t1f, t2f) : 
    v1i, v1f, v2f = round(L1/t1i, 4), round(L1/t1f, 4), round(L2/t2f, 4)
    pi = round(m1*v1i, 4)
    if m1 < m2 : 
        pf = round((m2*v2f) - (m1*v1f), 4)
    else : 
        pf = round((m1*v1f) + (m2*v2f), 4)
    Ki = kinetic_energy(m1, v1i)
    Kf = kinetic_energy(m1, v1f) + kinetic_energy(m2, v2f)
    p_error_rate = error_rate(pi, pf)
    Ek_error_rate = error_rate(Ki, Kf)
    return "결과 = v1i: {}, v1f: {}, v2f: {}, Pi: {}, Pf: {}, P의 오차율: {}, Ki: {}, Kf: {}, K의 오차율: {}".format(v1i, v1f, v2f,pi,pf,p_error_rate,Ki,Kf,Ek_error_rate)

# 실험 B 함수
def inelastic_collision(m1, m2, t1i, tf) : 
    v1i = round(L1/t1i, 4)
    pi = round(m1*v1i, 4)
    vf_theory = round(pi/(m1+m2), 4)
    vf_ex = round(L2/tf, 4)
    pf_theory, pf_ex = round((m1+m2)*vf_theory, 4), round((m1+m2)*vf_ex, 4)
    Ki = kinetic_energy(m1, v1i)
    Kf_theory, Kf_ex = kinetic_energy(m1+m2,vf_theory), kinetic_energy(m1+m2,vf_ex)
    p_error_rate = error_rate(pi,pf_ex)
    K_error_rate = error_rate(Kf_theory-Ki, Kf_ex-Ki)
    return "결과 = v1i: {}, vf의 이론값: {}, vf의 실험값: {}, Pi: {}, Pf의 이론값: {}, Pf의 실험값: {}, P의 오차율: {}, Ki: {}, Kf의 이론값: {}, Kf의 실험값: {}, K의 오차율: {}".format(v1i, vf_theory, vf_ex, pi, pf_theory, pf_ex,p_error_rate, Ki, Kf_theory, Kf_ex, K_error_rate) 

# 변하지 않는 것
L1 = float(input("글라이더1의 플래그 길이를 입력하세요(단위:m): "))
L2 = float(input("글라이더2의 플래그 길이를 입력하세요(단위:m):"))

a = input("탄성충돌 실험이면 a, 완전비탄성충돌 실험이면 b를 입력하세요: ")
# A. 탄성충돌 실험
if a=='a' :
    m1 = float(input("글라이더1의 질량을 입력하세요(단위:kg): "))
    m2 = float(input("글라이더2의 질량을 입력하세요(단위:kg): "))
    t1i = float(input("충돌 전 글라이더1이 포토게이트를 지나는 시간을 입력하세요: "))
    t1f = float(input("충돌 후 글라이더1이 포토게이트를 지나는 시간을 입력하세요: "))
    t2f = float(input("충돌 후 글라이더2가 포토게이트를 지나는 시간을 입력하세요: "))
    print(elastic_collision(m1, m2, t1i, t1f, t2f))
    
# B. 완전비탄성 충돌실험
elif a=='b' : 
    m1 = float(input("글라이더1의 질량을 입력하세요(단위:kg): "))
    m2 = float(input("글라이더2의 질량을 입력하세요(단위:kg): "))
    t1i = float(input("충돌 전 글라이더1이 포토게이트를 지나는 시간을 입력하세요: "))
    tf = float(input("충돌 후 글라이더가 포토게이트를 지나는 시간을 입력하세요: "))
    print(inelastic_collision(m1, m2, t1i, tf))
```

```python
# 관성모멘트와 각운동량 보존 실험

print("실험A 관성모멘트 I 측정")
# 일정한 것
m = float(input("추와 추걸이의 질량 합을 입력하세요(단위 kg): "))
r = float(input("회전축의 반지름 r을 입력하세요(단위 m): "))

# 추의 가속도 함수
def acceleration(r, w, t) : 
    a = round((r*w)/t, 4)
    return a

# 장력 T 함수
def tension(m, a) : 
    g = 9.80665
    return round(m*(g-a), 4)

# 관성모멘트 함수
def inertia(m, r, a) : 
    g = 9.80665
    I = round(m*pow(r,2)*((g-a)/a), 8)
    return I

# A. 관성모멘트 측정
j = 0
I_lst = []
while j <= 1 : 
    a_lst = []
    k = 0
    while k<=4 : 
        w = float(input("각속도 w를 입력하세요: "))
        t = float(input("낙하시간 t를 입력하세요: "))
        a = acceleration(r,w,t)
        T = tension(m, a)
        a_lst.append(a)
        print("\n가속도: {}, 장력: {}\n".format(a, T))
        k += 1
    a_average = round(sum(a_lst)/5, 5)
    I = inertia(m,r,a_average)
    print("관성모멘트 I: " + str(I))
    I_lst.append(I)
    j += 1

If, Ii = I_lst[0], I_lst[1]

# B. 각운동량 보존
def conservation_of_angular_momentum(wi, wf, Ii, If) : 
    Li, Lf = round(Ii*wi, 4), round(If*wf, 4)
    L_error_rate = error_rate(Li, Lf)
    Ki, Kf = kinetic_energy(Ii, wi), kinetic_energy(If, wf)
    Kfth = round(Ki*Ii/If, 4)
    K_error_rate = error_rate(Kfth, Kf)
    return "\n결과 = Li: {}, Lf: {}, 각운동량 오차율: {}, Ki: {}, Kf: {}, Kfth: {}, 회전운동에너지 오차율: {}". format(Li,Lf,L_error_rate,
                                                                                                 Ki,Kf,Kfth,K_error_rate)
print('')
print("실험B 사각질량의 각운동량 보존")
j = 0
while j<=4 : 
    wi = float(input("사각질량이 20cm일 때의 각속도 wi를 입력해주세요: "))
    wf = float(input("사각질량이 10cm일 때의 각속도 wf를 입력해주세요: "))
    result = conservation_of_angular_momentum(wi, wf, Ii, If)
    print(result)
    j+= 1

# C. 디스크와 링
print('')
print('실험C 디스크와 링의 각운동량 보존')
Md  = float(input("디스크의 질량을 입력해주세요(단위 kg): "))
R = float(input("디스크의 반지름을 입력해주세요(단위 m): "))
Mr = float(input('링의 질량을 입력해주세요(단위 kg): '))
R1 = float(input('링의 내경 R1을 입력해주세요(단위 m): '))
R2 = float(input('링의 외경 R2를 입력해주세요(단위 m): '))
Id, Ir = 0.5*Md*pow(R, 2), 0.5*Mr*(pow(R1,2)+pow(R2,2))
print('\n디스크의 관성모멘트: {}, 디스크와 링의 관성모멘트: {}\n'.format(Id, Id+Ir))

j = 0
while j<=4 : 
    wi = float(input("사각질량이 20cm일 때의 각속도 wi를 입력해주세요: "))
    wf = float(input("사각질량이 10cm일 때의 각속도 wf를 입력해주세요: "))
    result = conservation_of_angular_momentum(wi, wf, Id, Id+Ir)
    print(result)
    j+= 1
```

```python
# 단진자 운동

def period(l) : 
    from math import pi
    g = 9.80665
    return round(2*pi*pow(l/g, 0.5), 4)

def gravity_theory(T, l) : 
    from math import pi
    return round(4*pow(pi, 2)*l/pow(T, 2), 4)

g = 9.80665
l = float(input("실의 길이를 입력하세요(단위 m): "))
T_ex = float(input("주기 T의 측정값을 입력하세요: "))
T_theory = period(l)
T_error_rate = error_rate(T_theory, T_ex)
g_ex = gravity_theory(T_ex,l)
g_error_rate = error_rate(g, g_ex)
print("결과 = T의 이론값: {}, T의 오차율: {}, g의 실험값: {}, g의 오차율: {}(단, g = 9.80665m/s^2)".format(T_theory,T_error_rate,
                                                                                       g_ex,g_error_rate))
```
